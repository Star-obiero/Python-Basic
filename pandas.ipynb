Using .map() to transform values

```python
pclass_mapping={
    '3':'least expensive',
    '2':'standard',
    '1':'most luxirious & expensive'
}
pclass_mapping
```

```python
titanic_df['Pclass'].map(pclass_mapping)
```

```python
titanic_df['Pclass']=titanic_df['Pclass'].map(pclass_mapping)
titanic_df['Pclass']
```

Passing in a function

```python
def class_fare(fare):
          if fare>=12:
               return True
          else:
               return False
```

```python
titanic_df['Fare'].map(class_fare)
```

Rather than changing lets creat a new column to hold this changes

```python
titanic_df['farre_per_class']=titanic_df['Fare'].map(class_fare)
titanic_df['farre_per_class']
```
